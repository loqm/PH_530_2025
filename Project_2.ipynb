# Abstract
We propose a decoy-pulse method to overcome the photon-number-splitting attack for Bennett-
Brassard 1984 quantum key distribution protocol in the presence of high loss: A legitimate user
intentionally and randomly replaces signal pulses by multiphoton pulses (decoy pulses). Then they
check the loss of the decoy pulses. If the loss of the decoy pulses is abnormally less than that of signal
pulses, the whole protocol is aborted. Otherwise, to continue the protocol, they estimate the loss of
signal multiphoton pulses based on that of decoy pulses. This estimation can be done with an assumption
that the two losses have similar values.
# Introduction
### Quantum Information Processing  
Quantum systems enable certain computational and cryptographic tasks that are infeasible for classical counterparts. These include:  
- Quantum cryptography  
- Quantum computation  
- Quantum metrology  

Beyond their practical significance, these capabilities also have profound theoretical and philosophical implications.  

### The BB84 Quantum Key Distribution (QKD) Protocol  
Among various quantum information processing protocols, the Bennett-Brassard 1984 (BB84) quantum key distribution (QKD) protocol stands out as one of the most promising. It is anticipated to be the first practical implementation of quantum information processing.  

### Challenges in QKD Implementation  
Despite its potential, the practical realization of QKD for global secure communications faces significant challenges:  

### Distance Limit  
- QKD implementation has been successfully demonstrated over distances of tens of kilometers.  
- Extending this range remains a critical challenge.  

### Noise and Decoherence  
- Like classical signals, quantum signals are susceptible to noise and decoherence, impacting their integrity over long distances.  

## Overcoming Distance Limitations  
### Quantum Repeaters  
- A natural approach to extending QKD distances is amplifying quantum signals at intermediate locations.  
- However, the no-cloning theorem prevents straightforward amplification of quantum signals.  
- Quantum repeaters offer a potential solution, allowing quantum signals to be transmitted through noisy environments.  
- Unfortunately, current technology makes quantum repeaters difficult to implement.  

### Alternative Approach: Surface-to-Satellite Free-Space QKD  
- One of the most promising alternatives is surface-to-satellite free-space BB84 QKD.  
- However, this method suffers from high photon loss.  
- High loss poses a severe risk to the BB84 protocol, making it vulnerable to photon-number splitting (PNS) attacks.  

### Hence our purpose is to propose the decoy-pulse method to overcome the PNS attack for BB84 protocol in the presence of high loss.

# PNS Attack
### Assumptions  
- Alice, one of the legitimate participants, uses a photon source with the following emission properties:  
  - A single-photon pulse with a 90% probability.  
  - A multiphoton pulse with a 10% probability.  
- Multiphoton pulses are generated inadvertently, and their emission timing is unknown.  
- The channel has a loss factor $l$  of 90%, meaning it has a yield $y$ of 10% $y = 1 - l$ .  
- Bob, the other legitimate participant, uses practical detectors that are insensitive to photon numbers.  

### Photon-Number Splitting (PNS) Attack by Eve  
Eve, the eavesdropper, employs the following attack strategy:  
1. She measures the number of photons in each pulse.  
2. If the pulse contains a single photon, she blocks it.  
3. If the pulse contains multiple photons, she splits them:  
   - She retains one photon.  
   - The remaining photons are sent to Bob through an ideal lossless channel.  
4. Eve is assumed to have unlimited technological and computational power, constrained only by the laws of nature.  
5. Bob receives only 10% of the photon pulses, which aligns with the expected yield.  

### Security Implications  
- Eve can extract full information about the key by measuring the preserved photons using the publicly announced measurement basis.  
- Under the best-case assumption for Eve, all multiphoton pulses are exploited for PNS attacks.  
- If the yield $ y $ is less than the probability $  p_{\text{multi}}  $ of multiphoton generation, the scheme becomes completely insecure against PNS attacks. This is because Bob then has no means to detect any interference by Eve.
- Therefore, for the BB84 protocol to remain secure, the yield $ y $ must be greater than $ p_{\text{multi}} $. <br>
$$y > p_{\text{multi}}$$
### Requirements for Secure QKD  
- When the yield $y$ is very low, an almost perfect single-photon generator is required.  
- This requirement has driven the development of single-photon sources and their use in QKD implementations.  
- However, perfect single-photon sources are not practically achievable.  

### Surface-to-Satellite vs. Surface-to-Surface BB84 Protocol  
- In the case of surface-to-surface free-space BB84, the single-photon condition is nearly satisfied.  
- Surface-to-satellite QKD suffers from higher loss due to:  
  - The satellite moving rapidly.  
  - The increasing distance between communicating parties.  
- The security of a satellite-based protocol depends on the physical security of the satellite.  
  - It is assumed that Eve cannot secretly observe inside the satellite.  
  - However, justifying this assumption is challenging.  

### Reflection-Based Protocol  
- A proposed solution is to use a (possibly geostationary) satellite equipped with mirrors to connect surface users.  
- A reflection-based QKD protocol has been explored in multi-user QKD contexts.  
- However, this method introduces even higher losses, making a secure implementation difficult with current technology.  

## Detecting the Photon-Number Splitting (PNS) Attack  

### Direct Detection Methods  
- One approach is to monitor photon-number statistics, which might be disturbed by Eveâ€™s PNS attack.  
- In the case of weak coherent states, a simple PNS attack alters photon-number statistics, allowing Bob to detect it.  
- However, Eve can employ a more sophisticated PNS attack that preserves photon-number statistics, making detection harder.  
- Thus, additional countermeasures are required.

# Decoy-Pulse Method

### Main Idea
- The core idea behind the decoy-pulse method is:  
  - In a PNS attack, Eve selectively transmits multiphotons to Bob.  
  - As a result, the yield of multiphoton pulses becomes abnormally higher than that of single-photon pulses.  
- To counteract this, Alice intentionally and randomly replaces some signal photon pulses with multiphoton decoy pulses.  
- Since Eve cannot distinguish between multiphoton pulses from the signal source and those from the decoy source, their yields should be similar.  
- Alice and Bob can detect a PNS attack by checking the yield of the decoy source.


### Photon Number Representation  
- Consider a source that emits a pulse $\lvert n \rangle$ containing $n$ photons (all in the same polarization state) with probability $p_n$.  
- Here, $n = 0, 1, 2, \dots$, and the probabilities satisfy the normalization condition:  

  $$\sum_n p_n = 1$$  

- Each pulse is used to encode one bit of the key.  

### Coherent State Representation  
- Consider a source that generates a coherent state $\lvert \mu e^{i\theta} \rangle$.  
- By randomizing the phase $\theta$, the state reduces to a mixed state:  

  $$\rho = \int \frac{d\theta}{2\pi} \lvert \mu e^{i\theta} \rangle \langle \mu e^{i\theta} \rvert$$  

- This mixed state is equivalent to a mixture of Fock states:  

  $$\sum_n P_n(\mu) \lvert n \rangle \langle n \rvert$$  

  where $P_n(\mu)$ follows a Poissonian distribution:  

  $$P_n(\mu) = \frac{e^{-\mu} \mu^n}{n!}$$  

### Interpretation  
- A source emitting pulses in coherent states $\lvert \mu e^{i\theta} \rangle$ is, after phase randomization, equivalent to a source that emits an $n$-photon state $\lvert n \rangle$ with probability $P_n(\mu)$.

### Purpose of Signal and Decoy Sources  
- Alice adopts two photon sources:  
  - **Signal source** ($S$): Used to distribute the key.  
  - **Decoy source** ($S'$): Used to detect the Photon-Number Splitting (PNS) attack.  

- In the most practical case, both sources generate **coherent states**:  
  - For the signal source $S$, we set $\mu < 1$ so that it **mostly emits single-photon pulses**.  
  - For the decoy source $S'$, we set $\mu' \geq 1$ so that it **mostly emits multiphoton pulses**.  

- The polarization of the pulses from the decoy source is randomized so that, as long as the photon numbers match, they **cannot be distinguished from the signal source**.  

### Detection Assumptions  
- Bob uses **practical photon detectors** that are **insensitive to photon numbers**.  
- Define the **yields**:  
  - $y_n$: Relative frequency with which an $n$-photon pulse from the signal source is registered by Bobâ€™s detector.  
  - $y'_n$: Relative frequency with which an $n$-photon pulse from the decoy source is registered by Bobâ€™s detector.  
  - Here, $0 \leq y_n, y'_n \leq 1$.  

- **Important observation:**  
  - The yields can be **unity** even if some photons in a pulse are lost because Bobâ€™s detector does **not count lost photons**.  

### Yield Formulas  
- The **yield of the signal source** ($Y_s$) and the **yield of the decoy source** ($Y_d$) are given by:  

  $$
  Y_s = \sum_{n=0}^{\infty} P_n(\mu) y_n
  $$  

  $$
  Y_d = \sum_{n=0}^{\infty} P_n(\mu') y'_n
  $$  

  where $P_n(\mu)$ and $P_n(\mu')$ are the Poisson distributions of the signal and decoy sources, respectively:

  $$
  P_n(\mu) = \frac{e^{-\mu} \mu^n}{n!}, \quad P_n(\mu') = \frac{e^{-\mu'} \mu'^n}{n!}
  $$  


- The yields $Y_s$ and $Y_d$ **can be directly detected** by Bob.  
- We also define the **yield of only multiphoton pulses** from the signal source, denoted as $Y_m^s$, given by:

  $$
  Y_m^s = \sum_{n \geq 2} P_n(\mu) y_n
  $$   
- The quantity $Y_m^s$ **cannot be directly measured** but can be **bounded** using other yields.  
- We define the **normalized yield of multiphoton pulses** from the signal source, denoted as $\tilde{Y}_m^s$, given by:

  $$
  \tilde{Y}_m^s = \frac{\sum_{n \geq 2} P_n(\mu) y_n}{\sum_{n \geq 2} P_n(\mu)}
  $$  

- This represents the **fraction of the total signal yield that comes from multiphoton pulses**.  


# Protocol

### Step 1: Performing the BB84 Protocol
- Alice performs the BB84 protocol using the signal source $S$.
- She randomly replaces the signal source $S$ with the decoy source $S'$ with a probability $\alpha$.

### Step 2: Announcement and Public Discussion
- Bob announces that he has received all photon pulses.
- Alice then publicly announces which pulses were from the decoy source.
- Alice and Bob estimate:
  - The total yield of the signal source, $Y_s$.
  - The total yield of the decoy source, $Y_d$.

### Step 3: Aborting the Protocol if Necessary
- If $Y_d$ is significantly larger than $Y_s$, they abort the entire protocol.
- Otherwise, they proceed to estimate the yield of multiphoton pulses from the signal source.

### Step 4: Estimating the Multiphoton Pulse Yield
- The yield of multiphoton pulses from the signal source, $Y_m^s$, is estimated using the yield of the decoy source, $Y_d$.

### Step 5: Key Observation on Eveâ€™s Attack
- Eve cannot distinguish multiphoton pulses from the signal source and the decoy source.
- The normalized yield of multiphoton pulses from the signal source should be similar to that of the decoy source.
- Mathematically, the yields must satisfy:

  $$
  y_n = y'_n
  $$  
- For a given pulse with a certain photon number, Eve cannot gain any more information about which source the pulse originated from beyond what she can infer using Bayesâ€™s law.
- The only way for Eve to exploit Bayesâ€™s law is by manipulating the values of $y_n$ as she pleases.
- We can hence infer from above equations that:
 $$
\sum_{n \geq 2} P_n(\mu') y_n \leq Y_d
  $$  

# Bounding $Y_m^s$

Now the problem is how $Y_m^s = \sum_{n=2}^{\infty} P_n(\mu) y_n$ is bounded by above equation.

- Eveâ€™s goal is to make $Y_m^s$ as large as possible, for a given yield of decoy source $Y_d$.
- In other words, she aims to maximize the ratio:
  $$
  A \equiv \frac{\sum_{n=2}^{\infty} P_n(\mu) y_n}{\sum_{n=2}^{\infty} P_n(\mu') y_n}
  $$
- Let us now note that, for $\mu <\muâ€²$ as we will choose,
$$
\frac{P_n(\mu)}{P_n(\mu')} > \frac{P_m(\mu)}{P_m(\mu')}, \quad \text{if } n < m
$$
- This relationship holds because for Poissonian statistics:

$$
\frac{P_n(\mu)}{P_n(\mu')} = \frac{e^{-\mu}\mu^n/n!}{e^{-\mu'}(\mu')^n/n!} = e^{\mu'-\mu} \left(\frac{\mu}{\mu'}\right)^n
$$

- The critical ratio $A$ representing Eve's optimal attack strategy is bounded as:

$$
A = \frac{\sum_{n=2}^\infty P_n(\mu)y_n}{\sum_{n=2}^\infty P_n(\mu')y_n} \leq \frac{P_2(\mu)}{P_2(\mu')}
$$
- It is because:


\begin{align*}
P_2(\mu)/P_2(\mu') - A &= \left[1 \bigg/ \left\{P_2(\mu')\sum_{n=2}^\infty P_n(\mu')y_n\right\}\right] \\
&\quad \times \left[P_2(\mu)\sum_{n=2}^\infty P_n(\mu')y_n - P_2(\mu')\sum_{n=2}^\infty P_n(\mu)y_n\right] \\
&= \left[1 \bigg/ \left\{P_2(\mu')\sum_{n=2}^\infty P_n(\mu')y_n\right\}\right] \\
&\quad \times \sum_{n=2}^\infty y_n\left\{P_2(\mu)P_n(\mu') - P_2(\mu')P_n(\mu)\right\} \geq 0
\end{align*}

- $y_n, P_n \geq 0$  
- Equality holds when $y_2 > 0$ and $y_i = 0$ for $i \geq 3$  


## Optimal Eavesdropping Strategy
- Focuses exclusively on 2-photon pulses ($y_2 > 0$)  
- Blocks all higher-photon pulses ($y_i = 0$ for $i \geq 3$)  

- Higher-photon pulses ($n \geq 3$) are more likely to originate from the decoy source  
- Eve minimizes detection risk by blocking these pulses  


## Final Bound
Using the bound of $A$, we obtain the following:
$$
Y_s^m \leq \frac{P_2(\mu)}{P_2(\mu')} Y_d
$$
The normalized multiphoton yield is bounded by:  
$$
\tilde{Y}_s^m \leq \frac{1}{P_2(\mu')} \frac{P_2(\mu)}{\sum_{n=2}^\infty P_n(\mu)} Y_d
$$

For practical parameters ($\mu=0.3$, $\mu'=1.0$):  
- $P_2(\mu')$ and $\frac{P_2(\mu)}{\sum_{n=2}^\infty P_n(\mu)}$ are order unity  
- $\tilde{Y}_s^m$ and $Y_d$ thus share similar magnitudes  
$$ \tilde{Y}_s^m \sim Y_d$$

# Condition for security
- In order that the protocol be secure, the total number of pulses that are detected must be greater than that of attacked ones.
- The number of attacked pulses is,
$$ \left[ \sum_{n=2}^{\infty} P_n(\mu) \right] \tilde{Y}_s^m $$
- Thus the condition $(y>p_{multi})$ reduces to:
$$
Y_s > max \{\left[ \sum_{n=2}^{\infty} P_n(\mu) \right] \tilde{Y}_s^m\}
$$
where the maximum is taken over all strategies by Eve.
- Combining the above equation with the bound on $\tilde{Y}_s^m$, the final condition for security becomes:
$$ Y_s > \frac{P_2(\mu)}{P_2(\mu')} Y_d $$

# Interpretation
- Let us estimate the quantities where both signal and decoy sources generate coherent states with Poissonian statistics:
  - Signal source photon number distribution: $P_n(\mu)$
  - Decoy source photon number distribution: $P_n(\mu')$

- In normal protocol operations (without Eve's interference):
  - The yield $Y_d$ will be larger than $Y_s$ by a factor of $\mu'/\mu$
  - The condition reduces to:
    $$
    \frac{P_2(\mu)}{P_2(\mu')} \frac{\mu'}{\mu} = \frac{e^{\mu'}}{\mu'} \frac{\mu}{e^{\mu}} < 1
    $$

- For a given $\mu'$, the condition is satisfied when $\mu$ is sufficiently small, Because $\mu/e^{\mu} < \mu$
- Example case:
     - When $\mu = 0.3$ and $\mu' = 1.0$
     - Left-hand side of the equation â‰ˆ 0.604

# Key Advantages
- One might say that the mean photon number of signal source, ðœ‡, should still be quite smaller than unity even with the decoy-pulse method and thus there is no improvement over the usual protocols without the decoy-pulse method.
- But The security condition does not contain a term that amounts to channel loss. Thus the condition for security can be satisfied no matter how high the loss is in the normal operations.

# Generalisation
- Applicable to sources with any probability distribution $p_n$
  - Example: Almost-perfect single-photon generator with:
    - Photon-number distribution:
      - $p_1 = 1 - \epsilon$
      - $p_i = k / l!$ for $i = 2, 3, \ldots$
    - Where:
      - $\epsilon \ll 1$
      - $k$ is a constant satisfying $\sum_{i=2}^{\infty} p_i = \epsilon$

- Security condition when using decoy source $S'$:
  $$
  Y_s > \frac{\epsilon}{P_2(\mu')} Y_d
  $$
  
- Key observations:
  1. Equation above is easily satisfied when $\epsilon$ is sufficiently small
  2. The key generation rate:
     - Proportional to the margin of satisfaction
     - Can be larger than when using weak-coherent states as the signal source

### General principle:
 The decoy-pulse method is more efficient when the photon-number statistics of signal and decoy sources are more similar

# Simulation Setup
### Parameters
- $\mu = 0.3$: Mean photon number for signal pulses.  
- $\mu' = 1.0$: Mean photon number for decoy pulses.  
- $\eta = 0.1$: Channel transmission efficiency (10% survival).  
- $n_{\text{max}} = 10$: Maximum photons/pulse considered.
# Code Implementation

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Parameters
mu = 0.3        # Signal source mean photon number
mu_prime = 1.0  # Decoy source mean photon number
eta = 0.1       # Channel efficiency (10% transmission)
n_max = 10      # Max photons considered

# Poisson distribution
def poisson(n, mu):
    return np.exp(-mu) * (mu**n) / math.factorial(n)

# Generate photon number probabilities
n_values = np.arange(0, n_max + 1)
P_signal = np.array([poisson(n, mu) for n in n_values])
P_decoy = np.array([poisson(n, mu_prime) for n in n_values])

# Yield calculation function
def calculate_yield(P, eta, attack=False):
    y = np.array([1 - (1 - eta)**n for n in n_values])
    if attack:
        # Block single-photon pulses (n=1)
        y[1] = 0
    return np.sum(P * y)

# Calculate yields for both scenarios
Y_s_normal = calculate_yield(P_signal, eta)
Y_d_normal = calculate_yield(P_decoy, eta)
Y_s_attack = calculate_yield(P_signal, eta, attack=True)
Y_d_attack = calculate_yield(P_decoy, eta, attack=True)
```

# Results and Visualization

```python
# Plotting
plt.figure(figsize=(10, 6))

# Normal operation
plt.subplot(1, 2, 1)
plt.bar(['Signal (Yâ‚›)', 'Decoy (Yâ‚ˆ)'], [Y_s_normal, Y_d_normal], color=['blue', 'orange'])
plt.title('Normal Operation\n(No Eve)')
plt.ylabel('Yield')
plt.ylim(0, 0.35)

# Under attack
plt.subplot(1, 2, 2)
plt.bar(['Signal (Yâ‚›)', 'Decoy (Yâ‚ˆ)'], [Y_s_attack, Y_d_attack], color=['blue', 'orange'])
plt.title('Under PNS Attack\n(With Eve)')
plt.ylim(0, 0.35)

plt.tight_layout()
plt.show()

# Print quantitative results
print(f"[Normal] Yâ‚› = {Y_s_normal:.3f}, Yâ‚ˆ = {Y_d_normal:.3f}")
print(f"[Attack] Yâ‚› = {Y_s_attack:.3f}, Yâ‚ˆ = {Y_d_attack:.3f}")
print(f"Signal yield drop: {100*(Y_s_normal - Y_s_attack)/Y_s_normal:.1f}%")
print(f"Decoy yield drop: {100*(Y_d_normal - Y_d_attack)/Y_d_normal:.1f}%")
```

# Interpretation
  
- *Under attack*: The fall is $Y_s$ is 75.2% in the presence of Eve. Under normal circumstances, $Y_d$ must have also suffered a similar fall in value. But due to the presence of Eve, which focuses on sending Multiphoton pulses to Bob, the value of $Y_d$ boosts up and hence the reduction is only 38.7%. This alerts Aloce and Bob and they can drop the communication.  
- This demonstrates the effectiveness of the decoy-pulse method in detecting PNS attacks under high loss.
