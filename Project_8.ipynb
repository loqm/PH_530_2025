# **1. Introduction**

Quantum information technologies offer advantages like secure communication, precise measurements, and efficient computing. However, quantum states are fragile and easily disturbed by environmental interactions. To enhance capacity and security, systems often use higher-dimensional Hilbert spacessuch as those defined by spatial photon modes (e.g., orbital angular momentum). While these offer more information encoding possibilities, they also suffer from losses and noise, especially in multiphoton systems.

Traditionally, analyzing such systems is difficult due to their complexity. Some methods model them using discrete spatial modes, which work well for Gaussian states but dont scale easily. A promising alternative is to use phase-space methods, like the Moyal formulation of quantum mechanics, which represents states as functions (not operators) on phase space. This allows tools like the Wigner distribution to describe quantum systems more intuitively.

However, incorporating spatial and particle-number degrees of freedom into this formulation has been a challenge. Recent work introduced spatiotemporal quadrature bases, which unify these degrees of freedom into a single, comprehensive Hilbert space. This paper builds on that idea to fully develop a functional phase space formalism for quantum optics.

We aim to:
1. Extend the standard Wigner distribution to a Wigner functional that includes both spatiotemporal and particle-number degrees of freedom.
2. Show how quantum optical states and operators can be analyzed using this framework.
3. emonstrate practical use cases, including the computation of Wigner functionals for well-known quantum states (like Fock and coherent states).

While the mathematics involves functional integrals (often complex), we show that many cases can be made tractable using Gaussian functionals and generating functionals. This formalism not only generalizes existing phase space approaches but also connects with methods in quantum field theory.

## 2.1 Eigenstates of Qaudrature Operators

Quadrature operators allows us for phase space formulation of quantum optics (analogous to position and momentum) which is essential for Wigner Functionals. Quadrature operator eigenvalue equations can be written as:

\begin{equation}
  q_s(k)|q = q_s(k)|q\\
  p_s(k)|p = p_s(k)|p \tag{1}
\end{equation}

Here, k represents the three-dimensional wave vector and the subscript s is the spin index. Now, creation operator ${^}_s(k)$ and annihilation operator $a_s(k)$ (mode operators corresponding to plane waves) used for the quantization of the electromagnetic field:

\begin{equation}
  q_s(k) = \frac{1}{\sqrt{2}}(a_s(k) + {^}_s(k))\\
  p_s(k) = -\frac{i}{\sqrt{2}}(a_s(k) - {^}_s(k)) \tag{2}
\end{equation}

Also they satisfy commutation relation $[q_s(k),p_s(k)] = i$. In standard 2nd quantization formalism the EM field operator (whicvh satisfy standard eqal-time commutation relation for bosonic (photons) field) can be written as:

\begin{equation}
  _(x) = _s \int \frac{d^3k}{(2\pi)^3\sqrt{2\omega_k}}[a_s(k){_}^s e^{-ikx} + {^}_s(k){_}^{s*} e^{ikx}] \tag{3}
\end{equation}

The prefactors ensure relativistic normalization. ${_}^s$ are polarization 4-vector satisfyiong gauge constraints and exponential terms represent plane waves. Now from quantum Field Theory, field operator fundamental commutation relation:

\begin{equation}
  [_(x),\hat{\textstyle \prod}_(y)] = ig_{}^{(3)}(x-y) \tag{4}
\end{equation}

Here, $\hat{\textstyle \prod}_(y)$ is conjugate momentum density in terms of E field. Now, if we expand these operators in terms of ladder operators, we get Lorentz covariant commutation relation:

\begin{equation}
  [a_s(k_1),{^}_r(k_2)] = (2\pi)^3\omega_1_{s,r}(k_1-k_2) \tag{5}
\end{equation}

Now, commutation relation for quadrature operators corrensponding to entire functions:

\begin{equation}
  [q_s(k_1),p_r(k_2)] = -\frac{i}{2}[(a_s(k_1) + {^}_s(k_2)),(a_r(k_1) - {^}_r(k_2))]\\
  [q_s(k_1),p_r(k_2)] = -\frac{i}{2}[-(2\pi)^3\omega_1_{s,r}(k_1-k_2)-(2\pi)^3\omega_1_{s,r}(k_1-k_2)]\\
  [q_s(k_1),p_r(k_2)] = i(2\pi)^3\omega_1_{s,r}(k_1-k_2) \tag{6}
\end{equation}

The real valued eigen functions $q_s,p_s\in L^2\{\mathbb{R}^3\}$ (Hilber space) meaning they form a complete basis and there are infinite many eigenvalues - so the choice is non-unique. But $q_s(k)$ globally represents $|q\rangle$ not jusr for individual k values. Therefore, the eigenvectors depend on eigenvalues not the wave vectors in infinite dimentional Hilbert space.

Now, we know $|vac\rangle$ is a state with no excitations satisfying relation $(k)|vac=0$ $k$.

\begin{equation}
    |q\rangle = \hat{a}_q^\dagger | \text{vac} \rangle, \quad
    |p\rangle = \hat{a}_p^\dagger | \text{vac} \rangle \tag{7}
\end{equation}

Now we define operators to obtain continous variable eigenstates:
\begin{equation}
    \hat{a}_Q^\dagger = \sqrt{2} \int \hat{a}^\dagger (k) q(k) dk
\end{equation}
\begin{equation}
    \hat{a}_P^\dagger = \sqrt{2} \int \hat{a}^\dagger (k) p(k) dk \tag{8}
\end{equation}
\begin{equation}
    \hat{a}_R^\dagger = \frac{1}{2} \int \hat{a}^\dagger (k) \hat{a}^\dagger (k) dk
\end{equation}

where $dk = \frac{d^3k}{\omega(2\pi)^3}$ where the last one is calles pair-creation ooperator, which does a squeezing like transformation that ensures correct normalization. The eigenstates of quadrature operators should be Gaussian-weighted exponentials of the ladder operators:

\begin{equation}
    \hat{a}_q^\dagger = \pi^{-\Omega/4} \exp \left(-\frac{1}{2} \|q\|^2 + \hat{a}_Q^\dagger - \hat{a}_R^\dagger \right)
\end{equation}
\begin{equation}
    \hat{a}_p^\dagger = 2^{\Omega/2} \pi^{\Omega/4} \exp \left(-\frac{1}{2} \|p\|^2 + i\hat{a}_P^\dagger + \hat{a}_R^\dagger \right) \tag{9}
\end{equation}

where, $\|f\|^2\equiv\int|f(k)|^2 dk$ for $f: \mathbb{R}^3 \to \mathbb{C}$ and also $ \Omega \equiv \int \delta(0) \, d^3 k$ clled the countable limit infinite volume factor which reprent cardinality of a countavly infinite et $\Omega = \aleph_0$ for infinite degrees of freedon in the quantum field theoretic setting.

## 2.2 Orthogonality and Completeness

The quadratures are fourier conjugate pairs which obey orthoganility conditions, expressed in terms of dirac  functionals:

\begin{equation}
  \langle q | q' \rangle = \delta[q - q']\\
  \langle p | p' \rangle = (2\pi)^\Omega \delta[p - p'] \tag{10}
\end{equation}

Here, the pre-factor $(2\pi)^\Omega$ comes due to fourier transform. In quantum field theory, eigenstates of quadrature operators form continous orthonormal basis, where:

\begin{equation}
  \delta[q - q'] = _k \delta[q(k) - q'(k)]  \tag{11}
\end{equation}

Since they form complete basis, the spatiotemporal quadrature bases are expressed in terms of functional integrals:

\begin{equation}
  \int |q\rangle \langle q| \mathcal{D}[q] = 1\\
  \int |p\rangle \langle p| \mathcal{D}^{\circ}[p] = 1 \tag{12}
\end{equation}

where, $\mathcal{D}^{\circ}[p] \equiv \mathcal{D} \left[ \frac{p}{2\pi} \right] = \frac{1}{(2\pi)^\Omega} \mathcal{D}[p]$ is the functional measure over all possible quadrature functions $p(k)$ and $q(k)$.

## 2.3 Notation for Contraction

The funtional integrals in previous section, can typicaly be evaluated when the integrand takes only Gaussian form. In that case, to simplify our notations and calculations, we introduce the -contraction scheme, where inner product between two functions or their integral over all possible wave vectors are written as:

\begin{equation}
\langle f, g \rangle \equiv \int f^*(k) g(k) \, d\bar{k} \equiv f^* \diamond g \tag{13}
\end{equation}

When the kernel $B(k,k')$ (representing the mixing between different wave vector components) is involved:

\begin{equation}
f^* \diamond B \diamond g \equiv \int f^*(k) B(k, k') g(k') \, d\bar{k} \, d\bar{k}' \tag{14}
\end{equation}

One example being the 2-point correlation dunction. But when only diagonal function $T(k)$ is involved, it can be represented in terms of simple multiplication operator:

\begin{equation}
f^* \diamond T \diamond g \neq \int f^*(k) T(k) g(k) \, d\bar{k} \tag{15}
\end{equation}

One example being the Feynmann Propagator.

## 2.4 Fourier Integrals

The overlap of 2 quadratures or kernel of fourier transform:
\begin{equation}
  \langle q | p \rangle = \exp \left[ i \int q(k) p(k) d\bar{k} \right] \equiv \exp(iq \diamond p) \tag{16}
\end{equation}

From equation 12:

\begin{equation}
  |p\rangle = \int dq |q\rangle \langle q | p \rangle = \int dq |q\rangle \exp(iq \diamond p) = |p\rangle = \int |q\rangle \exp(iq \diamond p) D[q]\\
  |q\rangle = \int dp |p\rangle \langle p | q \rangle = \int dp |p\rangle \exp(-iq \diamond p) = |q\rangle = \int |p\rangle \exp(-iq \diamond p) D^\circ[p] \tag{17}
\end{equation}

These are called Functional Fourier transform mappings which play a key role in convertion between two mutually unbiased quadratures. Now position representation of momentum operator and momentum representation of position operator is given by functional derivative, which's action is shown below:

\begin{equation}
  \frac{\delta}{\delta f(\mathbf{k})} f(\mathbf{k'}) = (2\pi)^2 \omega \delta(\mathbf{k} - \mathbf{k'}) \equiv 1 \tag{18}
\end{equation}

Now using equation (1),
\begin{equation}
  \hat{p}(k)|q\rangle = \int dq |q\rangle \langle q |\hat{p}(k)|q\rangle = \int dq|q\rangle \left[ -i \frac{\delta}{\delta q(k)} \right] = \int |q\rangle \left[ -i \frac{\delta}{\delta q(k)} \right] \mathcal{D}[q]\\
  \hat{q}(k)|p\rangle = \int dp |p\rangle \langle p |\hat{q}(k)|p\rangle = \int dp|p\rangle \left[ i \frac{\delta}{\delta p(k)} \right] = \int |p\rangle \left[ i \frac{\delta}{\delta p(k)} \right] \mathcal{D^\circ}[p] \tag{19}
\end{equation}

Using the above the qquadrature operators can be represented in dual basis:

\begin{equation}
  \hat{p}(k) = \int |q\rangle \left[ -i \frac{\delta}{\delta q(k)} \right] \langle q| \mathcal{D}[q]\\
  \hat{q}(k) = \int |p\rangle \left[ i \frac{\delta}{\delta p(k)} \right] \langle p| \mathcal{D}^{\circ}[p] \tag{20}
\end{equation}

Also, with the help of equation (20), Shift operators and their operations can be defined as:
\begin{equation}
  |q - q_0\rangle = \hat{T}(q_0) |q\rangle = \exp(i q_0 \diamond \hat{p}) |q\rangle\\
  |p - p_0\rangle = \hat{T}(p_0) |p\rangle = \exp(i \hat{p} \diamond p_0) |p\rangle \tag{21}
\end{equation}

## 2.5 Integrals for Dirac  functionals

Now, we have:

\begin{equation}
  \delta[q_1 - q_2] = \langle q_1 | q_2 \rangle = \int \langle q_1|p\rangle \langle p|q_2 \rangle \mathcal{D}^{\circ}[p] = \int \exp(iq_1 \diamond p - iq_2 \diamond p) \, \mathcal{D}^{\circ}[p] \tag{22}
\end{equation}

Similarly, for p states:

\begin{equation}
  \int \exp(-iq \diamond p_1 + iq \diamond p_2) \, \mathcal{D}[q] = (2\pi)^\Omega \delta[p_1 - p_2] \tag{23}
\end{equation}

Now, let us define $\alpha(k) = \frac{1}{\sqrt{2}}[q(k)+ip(k)]$ which can also serve as a parametric function. Then]:
\begin{equation}
  \alpha^*\diamond\alpha - \alpha^*\diamond\alpha_0 = \int \frac{1}{2}[(q_0-ip_0)(q+ip)-(q-ip)(q_0+ip_0)] d\bar k\\
   \alpha^*\diamond\alpha - \alpha^*\diamond\alpha_0 = i\int(q_0p-qp_0)d\bar k \tag{24}
\end{equation}

Now,
\begin{equation}
  \int \exp (\alpha^* \diamond \alpha - \alpha^* \diamond \alpha_0) \, \mathcal{D}^{\circ}[\alpha] = \int e^{i\int(q_0p-qp_0)d\bar k} \, \mathcal{D}[q] \mathcal{D}^{\circ}[p]\\
   \int \exp (\alpha^* \diamond \alpha - \alpha^* \diamond \alpha_0) \, \mathcal{D}^{\circ}[\alpha] = (2\pi)^\Omega [p_0] = (2\pi)^\Omega [\alpha_0] \tag{25}
\end{equation}

where,
\begin{equation}
  \mathcal{D}^{\circ}[\alpha] \equiv \mathcal{D}[q] \, \mathcal{D}^{\circ}[p]\\
  \delta[\alpha_0] \equiv \delta[q_0] \, \delta[p_0]
\end{equation}

## 2.6 Fixed-spectrum Fock states

The fixed spectrum fock state is defined by:

\begin{equation}
  |n_F\rangle = \frac{1}{\sqrt{n!}}{(a_F)^\dagger}^n |vac\rangle \tag{26}
\end{equation}

where, $(_F)^\dagger  \int ^\dagger (k)F(k)dk$ where fourier domain wave function $F(k)$ is defined as $||F||^2 = \int |F(k)|^2 dk = 1 $. Let us check if our simple commutation relation is valid:

\begin{equation}
  [a_F,{a_F}^\dagger] = [\int a(k)F^*(k)dk,\int a^\dagger (k')F(k')dk']\\
  [a_F,{a_F}^\dagger] = \int\int F^*(k)F(k')[a(k),a^\dagger (k')]dkdk'\\
  [a_F,{a_F}^\dagger] = \int\int F^*(k)F(k') (2\pi)^3 \omega \delta(\mathbf{k} - \mathbf{k'})\frac{d^3k}{(2\pi)^3} = \int F^*(k)F(k) dk\\
  [a_F,{a_F}^\dagger] = \int |F(k)|^2 dk = 1
\end{equation}

So, it is valid !! Now action of annihilation operator on fixed-spectrum Fock states is given by,

\begin{equation}
  a(k)|n_F\rangle = |(n-1)_F\rangle F(k)\sqrt{n} \tag{27}
\end{equation}

These states are eigen-states of the number operator:

\begin{equation}
  n \equiv \int {a}^(k)a(k)d k \tag{28}
\end{equation}

Now using equation (27):

\begin{equation}
  n|n_F\rangle = n|n_F\rangle \tag{29}
\end{equation}

## 2.7 Fixed-spectrum Coherent states

The Fixed-spectrum Coherent states is defined as:

\begin{equation}
  (k)|\alpha_F\rangle = \alpha(k)|\alpha_F\rangle \tag{30}
\end{equation}

 Fixed-spectrum coherent states are eigenstates of the annihilation operator $\hat{a}(k)$, with eigenvalue function $\alpha(k)$, a complex spectral function. They represent quantum states with well-defined spectral distributions and do not explicitly depend on the wave vector k. These states are widely used in quantum optics and condensed matter theory, particularly in Green function calculations. The subscript F indicates parameterization by a fixed spectrum rather than labeling a specific function.

 The associated displacement operator is defined as:

\begin{equation}
\hat{D}[\alpha_F] \equiv \exp \left( \alpha \diamond \hat{a}^\dagger - \alpha^* \diamond \hat{a} \right) \tag{31}
\end{equation}

The inner product between different fixed spectrum coherent states:

\begin{equation}
\langle \alpha_F | \beta_G \rangle = (e^{-\frac{|\alpha|^2}{2}} \sum_{n=0}^{\infty} \frac{{\alpha^*}^n}{\sqrt{n!}} \langle n|) (e^{-\frac{|\beta|^2}{2}} \sum_{m=0}^{\infty} \frac{\beta^m}{\sqrt{m!}} |m\rangle)\\
 \langle \alpha_F | \beta_G \rangle = \exp \left( -\frac{1}{2} \|\alpha\|^2 - \frac{1}{2} \|\beta\|^2 + \langle \alpha, \beta \rangle \right)\\
 \langle \alpha_F | \text{vac} \rangle = \exp \left( -\frac{1}{2} \|\alpha\|^2 \right) \tag{32}
\end{equation}

Athough not orthogonal, the fixed-spectrum coherent states resolve to identity in the Bragmann representation. So the completeness relation:

\begin{equation}
\mathbb{1} = \int |\alpha\rangle \langle \alpha| \, \mathcal{D}^\circ[\alpha] \tag{33}
\end{equation}

## 2.8 Quadrature representation of Coherent states

Now using the equation (8) definitions and properties in equation (30), we can show that:

\begin{equation}
\hat{a}_Q |\alpha_F \rangle = \sqrt{2} \int \hat{a} (k) q(k) dk |\alpha_F \rangle = |\alpha_F \rangle \sqrt{2} q \diamond \alpha_0\\
\hat{a}_R |\alpha_F \rangle = \frac{1}{2} \int \hat{a} (k) \hat{a} (k) dk |\alpha_F \rangle = |\alpha_F \rangle \frac{1}{2} \alpha_0 \diamond \alpha_0 \tag{34}
\end{equation}

Where, $\alpha_0$ is the complex-valued parameter function. Therefore:

\begin{equation}
\langle q | \alpha_F \rangle = \pi^{-\Omega/4} \langle \text{vac} | \exp \left( -\frac{1}{2} \|q\|^2 + \hat{a}_Q - \hat{a}_R \right) |\alpha_F \rangle\\
 \langle q | \alpha_F \rangle =  \pi^{-\Omega/4} \exp \left( -\frac{1}{2} \|q\|^2 - \frac{1}{2} \|\alpha_0\|^2 + \sqrt{2} q \diamond \alpha_0 - \frac{1}{2} \alpha_0 \diamond \alpha_0 \right) \tag{35}
\end{equation}

# **3. Wigner functional Theory**
## 3.1 Wigner function (Wigner quasiprobability distribution)

Its a quasiprobability distribution function (because it can be negative) was first introduce by Eugene Wigner in 1932 to make a link between quantum wave function and probability distribution in phase space.

The Wigner distribution W(x,p) of a pure state is defined as,
$$W(x,p)=\frac{1}{\pi } \int_{-\infty}^{+\infty} \psi^ (x+y) \psi(x-y)e^{2ipy/\hbar}dy \tag{1} $$

where x and p are psotion and momentum (could be any conjugate variable pair)

In general for any mix state the wigner function is the Wigner transform of the density matric,

$$W(x,p)=\frac{1}{\pi } \int_{-\infty}^{+\infty}<x-y|\hat\rho|x+y> e^{2ipy/\hbar}dy \tag{2}$$

One can easity get (1) from (2) by putting $\hat\rho=|\psi><\psi|$ for pure state. The wigner function only represent particle degrees of freedom.

## 3.2 Wigner functional theory

Now we discuss the formalism of this Wigner functional theory in Quantum optics.Without any loss of generality one can neglect the spin index and once necessary one can incorporates it into the formalism.

## 3.3 Definition of the Wigner functional:

The general definition of the wigner functional is,
$$W[q, p] \equiv \int \left\langle q + \frac{1}{2} x \middle| \hat{A} \middle| q - \frac{1}{2} x \right\rangle \exp(-ip \diamond x) \mathcal{D}[x] \tag{3}
$$
where $\hat{A}$ is the operator in Hilbart space for any quantum optical operator (including spin, spetial and temporal degree of freedom).Here square bracket indicates its a functional means the intigration depends of the functional form of the q and p. p and q are the quadrature of the state are real valued functions and one can combine them to a single complex valued function $\alpha=\frac{1}{\sqrt{2}}(q+ip)$ and Wigner function is represented by $W[\alpha]$.

The element of the density matric $\hat{\rho}$ is the density functional given by,
$$
\rho \left[ q + \frac{1}{2}x, q - \frac{1}{2}x \right] \equiv \left\langle q + \frac{1}{2}x \middle| \hat{\rho} \middle| q - \frac{1}{2}x \right\rangle.
\tag{4}$$
Now for pure state $\hat\rho=|\psi\left>\right<\psi|$ and $ \psi[\alpha]=\left<\alpha|\psi\right>$ we get,
$$\rho\left[q+\frac{1}{2}x,q-\frac{1}{2}x\right]=\psi\left[q+\frac{1}{2}x\right]\psi^\star\left[q-\frac{1}{2}x\right]\tag{5}$$
The Wigner functional represent both partcle number DOF and spatiotemporal degrees of freedom.

## 3.4 Weyl transformation
The Weyl transformation is the inverse of the Wigner transformation which transform the phase space probability into qunatum mechanical operator in Hilbert space.

Let consider $$q+\frac{1}{2}x=q''$$
   $$q-\frac{1}{2}x=q'\tag{6}$$

The reverse tranformation of (3) for density matrix $\rho$ gives density functional-
$$\rho\left[q+\frac{1}{2}x,q-\frac{1}{2}x\right]=\int W[q,p]e^{ip\diamond x}D^o[p]$$
$$\rho\left[q'',q'\right]=\int W\left[\frac{1}{2}(q''+q'),p\right]e^{ip\diamond x}D^o[p]\tag{7}.$$

The density matric written as,

$$
\hat{\rho} = \int \rho \left[ q + \frac{1}{2}x, q - \frac{1}{2}x \right] \left| q + \frac{1}{2}x \right\rangle \left\langle q - \frac{1}{2}x \right| D[q, x]
$$

$$
\hat{\rho} = \int W[q, p] e^{i p \diamond x} \left| q + \frac{1}{2}x \right\rangle \left\langle q - \frac{1}{2}x \right| D^{\circ}[p] D[q, x]
$$

Then the trace of the density matric is given by,

\begin{aligned}
\text{tr} \hat{\rho} &= \int W[q, p] e^{i p \diamond x}
\langle q | q + \frac{1}{2}x \rangle \langle q - \frac{1}{2}x | q \rangle
D^{\circ}[p] D[q, x] \\  
&= \int W[q, p] e^{i p \diamond x} \delta(x/2) \delta(-x/2) D^{\circ}[p] D[q, x] \\  
&= \int W[q, p] e^{i p \diamond x} \delta(x/2) D^{\circ}[p] D[q, x] \\  
&= \int W[q, p] D^{\circ}[p] D[q] \\  
&= \int W[\alpha] D^{\circ}[\alpha]  
\end{aligned}



This should be unity.

$$tr{\hat{\rho}}=\int W[\alpha] D^0[\alpha]=1 \tag{9}$$

The expectation value of an Matric corresponding to an observable is giben by the trace of the product of the density operator and the operator of the observable,

\begin{align}
\langle \mathbf{\hat{O}} \rangle &= \sum_i \rho_i \langle i | \mathbf{\hat{O}} | i \rangle \\
&= \sum_{i,j} \rho_i \langle i | \mathbf{\hat{O}} | j \rangle \langle j | i \rangle \\
&= \sum_{i,j} \rho_i \langle j | i \rangle \langle i | \mathbf{\hat{O}} | j \rangle \\
&= \sum_{i,j} \langle j | \rho_i | i \rangle \langle i | \mathbf{\hat{O}} | j \rangle \\
&= \text{tr}(\hat{\rho} \mathbf{\hat{O}})
\end{align}

In the integral form of the Wigner functional is represented by,

$$\left<\hat{O}\right>=tr[\hat{\rho} \hat{O}]=\int W_{\hat\rho}[\alpha]W_{\hat{O}}[\alpha]D^{0}[\alpha]$$

Where the wigner function for the operator $\hat{O}$ is given by,

$$W_{\hat{Q}}[\alpha] \equiv \int \left\langle q + \frac{1}{2} x \middle| \hat{O} \middle| q - \frac{1}{2} x \right\rangle \exp(-ip \diamond x) \mathcal{D}[x] \tag{3}
$$

## 3.5 Moyal product or star pruduct:

If we have two  Wigner functional $W_{A}$ and $W_{B}$ corresponding to two operator $\hat{A}$ and $\hat{B}$ respectlively then the WIgner function of the product operator $\hat{A}\hat{B}$ is given by,

$$
W_{\hat{A} \hat{B}}[q, p] = 2^{2\Omega} \int W_{\hat{A}}[q - q_1, p - p_1] W_{\hat{B}}[q - q_2, p - p_2] e^{[i 2(q_1 \diamond p_2 - q_2 \diamond p_1)] }\mathcal{D}[q_1, q_2] \mathcal{D}^{\circ}[p_1, p_2].$$
It represents the Moyal product or star product for Wigner functionals.

##3.6 Characteristic function


The density operator in the coherent state basis can be written as:

\begin{equation}
    \hat{\rho} = \int d^2\alpha \, P(\alpha) |{\alpha}\rangle \langle{\alpha}|.
\end{equation}
Here, $P(\alpha)$ is the Glauber-Sudarshan P-function, which represents the quantum state in terms of a classical probability distribution over coherent states.

The characteristic function is defined in terms of the Weyl displacement operator $D(\xi)$ as:

\begin{equation}
    \chi(\xi) = \text{Tr} \left[ D(\xi) \hat{\rho} \right]
\end{equation}

The density matrix $\hat{\rho}$ represents the quantum state.

The Weyl operator (also called the displacement operator) $D(\xi)$ is defined as:

\begin{equation}
    D(\xi) = e^{i x^T \Omega \xi}.
\end{equation}


where we define the phase quadratures for a single-mode system as:
$$
x^T = (x, p)
$$

and the symplectic matrix $\Omega$ as:

$$
\Omega =
\left[
\begin{array}{cc}
0 & 1 \\
-1 & 0
\end{array}
\right].
$$
The vector $\xi$ is given by:

\begin{equation}
\xi=
\left[
\begin{array} {cc}
\alpha \\
\beta
\end{array}
\right]
\end{equation}

So,
$$
\begin{equation}
   ix^{T}\Omega_{\xi} = i(\beta \hat{x} - \alpha \hat{p})
   = \frac{1}{\sqrt{2}} \left( -\alpha \hat{a} + i\beta \hat{a} + \alpha \hat{a}^\dagger + i\beta \hat{a}^\dagger \right)
\end{equation}
$$
Where, $\hat{a}$ and $\hat{a^{\dagger}}$ are the annhilation and creation operator.

where if we define $$ \eta = \frac{\alpha + i \beta}{\sqrt{2}} $$, we get

\begin{equation}
    i x^T \Omega_{\xi} = \eta \hat{a}^\dagger - \eta^* \hat{a}.
\end{equation}

So, the characteristic function can be written as

\begin{equation}
    \chi(\eta) = \text{tr} \left( \hat{\rho} e^{\eta \hat{a}^\dagger - \eta^* \hat{a}} \right).
\end{equation}

The characteristics function is the fourier transform of the PDF. Similarly here the characteristics functional is defined as the functional fourier transform of the  Wigner functional is given by,

$$\chi[\xi,\eta]=\int e^{ip  \eta - i\xi q}~W[q,p]~D[q]~D^{o}[p]$$

The one can obtained the Wigner functional from charasteristics by inverse functional fourier transform,

$$W[q,p]=\int e^{-(ip  \eta - i\xi q)}~\chi[\xi,\eta]~D[\eta]~D^{0}[\xi]$$

The Wigner functional for the Displacement operator is given by [As discussed later]

$$W[\alpha, \alpha_{0}]=e^{(\alpha^{\star}  \alpha_{0}-\alpha_{0}^{\star}  \alpha)}$$


##3.7 Marginal Probability

\begin{equation}
W[q,p] = \frac{1}{2\pi} \int e^{-i p q'} \langle q + \frac{1}{2} q' | \hat{\rho} | q - \frac{1}{2} q' \rangle \, D[q']
\end{equation}
As a quasi-probability distribution over the functional phase space, the Wigner functional of a state does not satisfy the criteria for a true probability density. However, one can derive a probability density from it by performing an integration over either p or q. By integrating the Wigner functional over p we can obtain,

\begin{equation}
\int W[q,p] D^\circ [p] \, = \frac{1}{2\pi} \int e^{-i p q'} \langle q + \frac{1}{2} q' | \hat{\rho} | q - \frac{1}{2} q' \rangle \, D[q'] D^\circ [p]
\end{equation}

Here,
\begin{equation}
\int dp \, e^{-i p q'} D^\circ [p] = \delta(q')
\end{equation}

Putting it into equation (3) we can find that

\begin{equation}
\int  \langle q - \frac{1}{2} q' | \hat{\rho} | q + \frac{1}{2} q' \rangle \delta[q'] \, D[q']
= \langle q|\hat{\rho}|q\rangle
\end{equation}
If we marginalize the function over p, then we will find the probability in the co-ordinate basis.

Similarly, if we integrate over q, then we have to insert the Identity operator in the p basis.


\begin{align}
\int D[q] W[q, p] &= \int\langle q + \frac{1}{2} q'| p_1 \rangle  \langle p_1 | \hat{\rho} | p_2 \rangle \langle p_2| q - \frac{1}{2} q' \rangle\exp(-i p \cdot q') D^\circ [p_1, p_2] D[q, q'] \\
& =  \langle p | \hat{\rho} | p \rangle
\end{align}

This represents probability in the p space.

#**4. Example of Wigner function**
##4.1 Wigner function of coherent state:
The charecteristic function is written as,
$\chi(\eta) = tr(\rho e^{\eta \hat{a}^{\dagger} - \eta^*\hat{a}})$.

The Wigner function of the coherent state is given by,
\begin{align*}
    W(\alpha) &= \frac{1}{\pi^2} \int e^{\eta^*\alpha - \eta\alpha^*}\chi_s(\eta) d^2\eta \\
    & = \frac{1}{\pi^2} \int e^{\eta^*\alpha - \eta\alpha^*} \langle \alpha_0 | \hat{D}(\eta) | \alpha_0 \rangle d^2\eta \\
    & = \frac{1}{\pi^2} \int e^{\eta^*\alpha - \eta\alpha^*} \langle 0 | \hat{D}^{\dagger}(\alpha_0) \hat{D}(\eta) \hat{D}(\alpha_0) | 0 \rangle d^2\eta \\
    & = \frac{1}{\pi^2} \int e^{\eta^*\alpha - \eta\alpha^*} e^{\frac{1}{2}(\alpha_0^*\eta-\eta^*\alpha_0)} \langle 0 | \hat{D}(-\alpha_0) \hat{D}(\alpha_0 + \eta) | 0 \rangle d^2\eta \\
    & = \frac{1}{\pi^2} \int e^{\eta^*\alpha - \eta\alpha^*} e^{\frac{1}{2}(\alpha_0^*\eta-\eta^*\alpha_0)} \langle 0 | \hat{D}(-\alpha_0) \hat{D}(\alpha_0 + \eta) | 0 \rangle d^2\eta \\
    & = \frac{1}{\pi^2} \int e^{\eta^*(\alpha-\alpha_0) - \eta(\alpha -\alpha_0)^*} e^{-\frac{|\eta|^2} {2}}d^2\eta \\
\end{align*}

Let $\eta = \eta_r + i\eta_i$ and $\alpha - \alpha_0 = (\alpha_r - \alpha_{0r}) + i(\alpha_i - \alpha_{0i})$

\begin{align*}
    &= \frac{1}{\pi^2} \int_{-\infty}^{\infty} e^{2i\eta_r(\alpha_i - \alpha_{i0})-\frac{1}{2}\eta_r^2 }d\eta_r \int_{-\infty}^{\infty} e^{-2i\eta_i(\alpha_r - \alpha_{0r})-\frac{1}{2}\eta_i^2 }d\eta_i
\end{align*}

Now using the standard integral,

\begin{equation*}
    \int_{-\infty}^{\infty} e^{-\frac{1}{2}ax^2 + iJx} dx = \left(\frac{2\pi}{a}\right)^{\frac{1}{2}} e^{\frac{-J^2}{2a}}
\end{equation*}

So, finally the Wigner Function of the Coherent state is given by,

\begin{equation}
    \boxed{W(\alpha) = \frac{2}{\pi} e^{-2|\alpha - \alpha_0|^2}}
\end{equation}  


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets
from IPython.display import display

# Grid definition
N = 100  # Number of points
alpha_Re = np.linspace(-7, 7, N)
alpha_Im = np.linspace(-7, 7, N)
X, Y = np.meshgrid(alpha_Re, alpha_Im)

# Wigner function of coherent state
def Wigner_fn(x, y, alpha_0):
    return (2/np.pi) * np.exp(-2 * np.abs((x + y * 1j) - alpha_0) ** 2)

# Function to update plot dynamically
def update_plot(alpha_0_Re, alpha_0_Im):
    alpha_0 = complex(alpha_0_Re, alpha_0_Im)

    # Calculate the Wigner function at each point
    W = Wigner_fn(X, Y, alpha_0)

    # Plot in 3D
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, W, cmap='viridis')

    # Labels and display
    ax.set_xlabel(r'$\Re(\alpha)$')
    ax.set_ylabel(r'$\Im(\alpha)$')
    ax.set_zlabel(r'$W(\alpha)$')
    ax.set_title(f'Wigner Function of a Coherent State ( = {alpha_0})')

    plt.show()

# Create sliders
alpha_0_Re_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=0, description=r'$\Re(\alpha_0)$')
alpha_0_Im_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=0, description=r'$\Im(\alpha_0)$')

# Display sliders and update function
interactive_plot = widgets.interactive(update_plot, alpha_0_Re=alpha_0_Re_slider, alpha_0_Im=alpha_0_Im_slider)
display(interactive_plot)
```

## 4.2 Displacement Oparator
In earlier discussion,  we have displacement operator ,
\begin{equation}
    \hat{D}[\alpha_F] = \exp(\alpha \diamond \hat{a}^\dagger -\alpha^*\diamond \hat{a)}
\end{equation}
Using Baker-Campbell-Hausdorff formula,
\begin{equation}
    \exp(\hat{X} + \hat{Y}) = \exp(\frac{-1}{2}[\hat{X} , \hat{Y}]) \exp(\hat{X}) exp(\hat{Y})
\end{equation}
here we have assumed $[\hat{X},[\hat{X},\hat{Y}]]= [\hat{Y},[\hat{X},\hat{Y}]] = 0$.


 So, we are getting,
$[\alpha \diamond \hat{a}^\dagger ,- \alpha^*\diamond \hat{a}]  = ||\alpha||^2 $

so, we have displacement oparator,

\begin{equation}
\boxed{
    \hat{D}[\alpha_0]  = \exp(\frac{-1}{2} ||\alpha_0||^2) \exp(\alpha \diamond \hat{a}^\dagger)\exp(- \alpha^* \diamond \hat{a})}
\end{equation}\
**Let's compute the overlap of coherent states $\langle \alpha_1 | \hat{D}[\alpha_0] | \alpha_2 \rangle $:**\


\begin{equation}
    \langle \alpha_1 | \hat{D}[\alpha_0] | \alpha_2 \rangle  = \langle \alpha_1 | \alpha_2 + \alpha_0 \rangle
\end{equation}
From earlier discussion, we have shown,
\begin{equation}
    \langle \alpha|\beta \rangle = \exp{(-\frac{|\alpha|^2 + |\beta|^2}{2} + \alpha^* \beta )}
\end{equation}
Using the same equation in our discussion,
\begin{equation}
       = \exp(\frac{-1}{2} ||\alpha_0||^2)  \exp(-\frac{|\alpha_1|^2 + |\alpha_2 + \alpha_0|^2}{2} + \alpha_1^* \diamond (\alpha_2 + \alpha_0))
\end{equation}
Simplifying we have,
\begin{equation}
    \langle \alpha_1 | \hat{D}[\alpha_0] | \alpha_2 \rangle  = \exp\left( -\frac{1}{2} |\alpha_0|^2 + \alpha_1^* \diamond \alpha_0 - \frac{1}{2} |\alpha_1|^2 - \frac{1}{2} |\alpha_2|^2 + \alpha_1^* \diamond \alpha_2 - \alpha_0^* \diamond \alpha_2 \right)
\end{equation}

We have the expression,
\begin{equation}
    \hat{W}_{\hat{A}}[\alpha] = N_0 \int \exp\left( -2|\alpha|^2 + 2 \alpha_1^* \diamond \alpha + 2 \alpha_2^* \alpha - \frac{1}{2} |\alpha_1|^2 - \frac{1}{2} |\alpha_2|^2 - \alpha_2^* \diamond  \alpha_1 \right) \langle \alpha_1 | \hat{A} | \alpha_2 \rangle \hat{D}^\circ[\alpha_1, \alpha_2] \,
\end{equation}
where $N_0$ is essentially $2^\Omega$.

Inserting $    \langle \alpha_1 | \hat{D}[\alpha_0] | \alpha_2 \rangle$ in the above equation , we obtain,
\begin{equation}
        \hat{W}_{\hat{D}}[\alpha] = N_0 \int \exp\left( -2|\alpha|^2 + 2 \alpha_1^* \diamond \alpha + 2 \alpha_2^* \alpha - \frac{1}{2} |\alpha_1|^2 - \frac{1}{2} |\alpha_2|^2 - \alpha_2^* \diamond  \alpha_1 \right)\langle \alpha_1 | \hat{D} | \alpha_2 \rangle \hat{D}^\circ[\alpha_1, \alpha_2] \,
\end{equation}


Putting the value of $\langle \alpha_1 | \hat{D} | \alpha_2  \rangle$,

\begin{equation}
\hat{W}_{\hat{D}} = N_0 \int e^ {\left( -2\|\alpha\|^2 + 2\alpha^* \diamond \alpha_1 + 2\alpha^*_2 \diamond \alpha - \|\alpha_1\|^2 -\|\alpha_2\|^2 - \alpha^*_2 \diamond \alpha_1 + \alpha^*_1 \diamond \alpha_0 + \alpha^*_1 \diamond \alpha_2 - \alpha^*_0 \diamond \alpha_2 - \frac{1}{2} \|\alpha_0\|^2 \right)}  D^\circ[\alpha_1, \alpha_2]
\end{equation}

Now we can figure out form of $W_{\hat{D}}[\alpha;\alpha_0]$ by integrating $\alpha_1$ and $\alpha_2$.

Bringing out all the $\alpha_1$ and $\alpha_2$ dependence,
\begin{equation}
 S = 2\alpha^* \diamond \alpha_1 + 2\alpha^*_2 \diamond \alpha - \|\alpha_1\|^2 -\|\alpha_2\|^2 - \alpha^*_2 \diamond \alpha_1 + \alpha^*_1 \diamond \alpha_0 + \alpha^*_1 \diamond \alpha_2 - \alpha^*_0 \diamond \alpha_2
\end{equation}
We know the famous gaussian integral
\begin{equation}
\int \exp\left(-\alpha^* \cdot K \cdot \alpha - \alpha^* \cdot \xi - \zeta^* \cdot \alpha \right) D^\circ[\alpha] = \frac{\exp\left( \zeta^* \cdot K^{-1} \cdot \xi \right)}{\det K}
\end{equation}
We can write the S in this matrix form,
\begin{equation}
S = -(\alpha_1^*, \alpha_2^*)  
\begin{bmatrix}  
1 & -1 \\  
1 & 1  \
\end{bmatrix}  
\begin{bmatrix} \alpha_1 \\ \alpha_2 \end{bmatrix}
+ (\alpha_1^*, \alpha_2^*)  
\begin{bmatrix}  
1 & 0 \\  
0 & 2  
\end{bmatrix} \begin{bmatrix} \alpha_0 \\ \alpha \end{bmatrix} + (\alpha^*, \alpha_0^*)  
\begin{bmatrix}  
2 & 0 \\  
0 & -1  
\end{bmatrix}
\begin{bmatrix} \alpha_1 \\ \alpha_2 \end{bmatrix}
\end{equation}
Comparing,
\begin{equation}
\alpha = \begin{bmatrix} \alpha_1 \\ \alpha_2 \end{bmatrix}, \xi = \begin{bmatrix} \alpha_0 \\ 2\alpha \end{bmatrix}, \zeta = \begin{bmatrix} 2\alpha \\ -\alpha_0 \end{bmatrix},K = \begin{bmatrix}  
1 & -1 \\  
1 & 1  
\end{bmatrix}
\end{equation}

```python
import sympy as sp
from IPython.display import display

# Define symbolic variables
alpha0, alpha = sp.symbols('alpha0 alpha')

# Define matrix K and vectors zeta, xi
K = sp.Matrix([[1, -1], [1, 1]])
xi = sp.Matrix([[alpha0], [2*alpha]])
zeta = sp.Matrix([[2*alpha], [-alpha0]])

# Compute inverse of K
K_inv = K.inv()
det_K = K.det()
#print("det of K:", det_K)
# Compute the desired expression: zeta*  K^(-1)  xi
result = (zeta.T.conjugate() * K_inv * xi)[0, 0]

# Simplify the expression
result = sp.simplify(result)
simplified_result = sp.simplify(sp.exp(result) / det_K)

# Display the symbolic result
print("Value of the integral =")
display(simplified_result)
```

<u>**Some Important Takeaways**</u>


*   Wigner function only incorporate particle-number degrees of freedom whereas Wigner functional incorporates both particle-number degrees of freedom and spatiotemporal degrees of freedom.

For example wigner function for coherent state is $W(\alpha;\alpha_0) = 2\exp(-2|\alpha-\alpha_0|^2) $ where as wigner functional for the full spectrum of coherent state is $W[\alpha;\alpha_0] = 2^\Omega \exp(-2||\alpha-\alpha_0||^2)$. In finction, \alpha 's are complex numbers and parameters whereas for functional they are complex funtions.
*   Normalising constatnts are also very different, the former is a finite number whereas for latter it associated with the cardinality of countable infinity.




## Time dependence of Wigner function for free particle
Say at t= 0,  we are starting with a gaussian wave packet, $\Psi(x,0) = A\exp(-\alpha x^2)$ for a free particle. From standred calculation of griffith's quantum mechanics,



\begin{equation}
    \Psi(x,t) = (\frac{2a}{\pi})^{1/4} \frac{\exp^{\frac{-\alpha x^2}{1+\frac{it}{\tau}}}}{\sqrt{1+\frac{it}{\tau}}}
\end{equation}



where $\tau = \frac{m}{2\hbar a}$ is the typical timescale of the problem.
\begin{equation}
    W(x,p,t) = \frac{1}{2\pi \hbar} \int dy \, e^{-ipy/\hbar} \Psi^*\left(x - \frac{y}{2}, t\right) \Psi\left(x + \frac{y}{2}, t\right)
\end{equation}



Now putting the value of $\Psi(x-\frac{y}{2}),\Psi(x+\frac{y}{2}) $, and integrating,


\begin{equation}
\boxed{
    W(x,p,t) = \frac{1}{\pi \hbar} e^{-\frac{p^2 (t^2 + \tau^2)}{2a \tau^2 \hbar^2} - 2ax^2 + \frac{2ptx}{\tau \hbar}} =
e^{ -2a \frac{(pt - mx)^2}{m^2} - \frac{p^2}{2a \hbar^2} }}
\end{equation}
Wigner function evolution follows for different time,

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets
from ipywidgets import interact

# Define constants
hbar = 1  # Set Planck's constant to 1 for simplicity
a = 1  # Some characteristic parameter
tau = 1  # Another characteristic parameter

def Wigner_function(x, p, t):
    return (1 / (np.pi * hbar)) * np.exp(
        - (p**2 * (t**2 + tau**2)) / (2 * a * tau**2 * hbar**2)
        - (2 * a * x**2)
        + (2 * p * t * x) / (tau * hbar)
    )

# Function to generate and show the plot
def plot_wigner_3d(t, elev=30, azim=45):
    x = np.linspace(-4, 4, 100)
    p = np.linspace(-4, 4, 100)
    x, p = np.meshgrid(x, p)
    W = Wigner_function(x, p, t)

    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, p, W, cmap='viridis', edgecolor='none')

    # Labels and title
    ax.set_xlabel("x")
    ax.set_ylabel("p")
    ax.set_zlabel("W(x, p, t)")
    ax.set_title(f"Wigner Function at t = {t}")

    # Enable interactive rotation
    ax.view_init(elev=elev, azim=azim)
    plt.show()

# Interactive widget
interact(plot_wigner_3d, t=(0, 50, 1), elev=(0, 90, 5), azim=(0, 360, 10));
```
