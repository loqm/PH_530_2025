# Optical Homodyne Detection:

## 1. Introduction

In quantum optics, Among various representations of quantum state of light, the **Wigner function** provides a powerful phase-space description of such states. To experimentally access the Wigner function, we can use **optical homodyne technique**, enabling measurement of field quadrature or **direct measurement of wigner function by photon counting**. To represent quantum states of light (e.g., vacuum state or coherent state) requires knowledge of **quadrature components** of the electromagnetic field. 

Optical homodyne detection offers one way to measure these quadratures.

---

## 3. Basic Principle of Optical Homodyne Detection

Homodyne detection is based on **interference** between:

- A **weak quantum signal** field $ \hat{a} $,
- A **strong coherent local oscillator (LO)** field $ \hat{a}_{\text{LO}} $.

The fields are mixed at a **50:50 beam splitter**, and the two outputs are detected by **balanced photodetectors**. The difference in photocurrents yields a measurement proportional to the **field quadrature**:

$$
\hat{X}_\theta = \frac{1}{2} \left( \hat{a} e^{-i\theta} + \hat{a}^\dagger e^{i\theta} \right)
$$

Here, $ \theta $ is the phase difference between the signal and LO. By scanning $ \theta $, one can measure different quadratures and reconstruct the full Wigner function through **quantum state tomography**.

---

## 4. Measurement Procedure

1. **Prepare signal and LO beams** from a coherent laser source.
2. Combine them at a **beam splitter**.
3. Use **balanced detectors** to measure photocurrent differences.
4. Vary the LO phase $ \theta $ using a piezo mirror or EOM.
5. Reconstruct the Wigner function using inverse Radon transform or maximum likelihood estimation.

---

## 5. Direct Photon-Counting Measurement of the Wigner Function

An alternative approach is to **directly measure the Wigner function** at a point in phase space using the formula:

$$
W(\alpha) = \frac{2}{\pi} \langle \hat{D}(\alpha) \hat{\Pi} \hat{D}^\dagger(\alpha) \rangle
$$

Here:
- $ \hat{D}(\alpha) $ is the **displacement operator**,
- $ \hat{\Pi} = (-1)^{\hat{n}} $ is the **parity operator**,
- $ \alpha $ is the complex phase-space coordinate.

This method involves:
- Displacing the quantum state by $ \alpha $ using a coherent probe beam,
- Counting photons to evaluate parity,
- Scanning $ \alpha $ to sample the Wigner function point-by-point.

---

## 6. Why Use Direct Photon Counting?

### 6.1. Simplicity and Directness
- Homodyne detection reconstructs the Wigner function **indirectly**, through post-processing of marginal distributions.
- Direct photon counting **measures it point-by-point** as an expectation value  no inversion algorithm is required.


### 6.2. Reduced Data Requirements
- Since each phase-space point is directly measured, fewer samples may be needed compared to the full tomographic scan of quadratures in homodyne detection.

---

## 7. Summary Comparison

| Feature                      | Homodyne Detection                  | Photon Counting (Direct)             |
|-----------------------------|-------------------------------------|--------------------------------------|
| Output                      | Quadrature distributions            | Wigner function at a point           |
| Reconstruction              | Required (e.g., inverse Radon)      | Not needed                           |
| Efficient for               | Gaussian states                     | Non-Gaussian, low-photon states      |
| Phase sensitivity           | Yes                                 | Yes                                  |

## Theoretical Background : Amandeep Singh

The Wigner function $ W(\alpha) $ is a fundamental concept in quantum optics and quantum information theory, offering a powerful representation of quantum states in a phase space similar to classical mechanics. Unlike classical probability distributions, the Wigner function can assume negative values, reflecting intrinsic quantum properties and nonclassical effects.

Formally, for a quantum state $ \hat{\rho} $, the Wigner function is defined as:

$$
W(\alpha) = \frac{2}{\pi} \text{Tr}\left[\hat{\rho} \hat{D}(\alpha)(-1)^{{n}}\hat{D}^\dagger(\alpha)\right]
$$

Expanding explicitly in terms of Fock states $ |n\rangle $, this becomes:

$$
\hat{W}(\alpha) = \frac{2}{\pi} \sum_{n=0}^{\infty} (-1)^n \hat{D}(\alpha) |n\rangle \langle n| \hat{D}^\dagger(\alpha)
$$

In this expression, $ |n\rangle $ denotes the photon number eigenstate (Fock state), and $ \hat{D}(\alpha) $ is the displacement operator.

### The Displacement Operator $ \hat{D}(\alpha) $

The displacement operator $ \hat{D}(\alpha) $ plays a crucial role in quantum optics. It is defined as:

$$
\hat{D}(\alpha) = e^{\alpha \hat{a}^\dagger - \alpha^* \hat{a}}
$$

Here, $ \hat{a} $ and $ \hat{a}^\dagger $ are the annihilation and creation operators, respectively. Physically, the displacement operator shifts a quantum state in the complex phase space by an amount given by $ \alpha $, which has amplitude and phase components. Mathematically, the displacement operation on a vacuum state creates a coherent state:

$$
\hat{D}(\alpha)|0\rangle = |\alpha\rangle
$$

Hence, coherent states $ |\alpha\rangle $ are the eigenstates of the annihilation operator $ \hat{a} $ and represent the quantum analog of classical harmonic oscillator states.

### Photon Number Parity and Wigner Function

The photon number parity operator is defined as:

$$
\hat{\Pi} = \sum_{n=0}^{\infty}(-1)^n |n\rangle \langle n|
$$

This operator distinguishes between states having even or odd numbers of photons. The expectation value of this parity operator, after displacing the state by $ \alpha $, directly gives the Wigner function:

$$
W(\alpha) = \frac{2}{\pi}\langle \hat{\Pi} \rangle_{\alpha} = \frac{2}{\pi}\sum_{n=0}^{\infty}(-1)^n p_n(\alpha)
$$

Here, $ p_n(\alpha) $ is the photon number distribution of the displaced quantum state. Experimentally, photon counting after displacement provides direct access to this distribution. By computing the alternating sum of photon-count probabilities, one obtains the Wigner function value at each phase-space point $ \alpha $.
# Relation between Experimental Setup and Wigner Function Measurement : Galabh Joshi

The goal is to relate the photon-counting statistics obtained from the proposed optical setup to the Wigner function of the signal field.

## 1. Output Mode Definition

A beam splitter with transmissivity \( T \) mixes the signal and probe modes. The annihilation operator of the output mode is:

$$
\hat{a}_{\text{out}} = \sqrt{T} \hat{a}_S - \sqrt{1 - T} \hat{a}_P
$$

Assuming the probe mode is in a coherent state $$ \ket{\alpha} $$, we can use $$\hat{a}_P \ket{\alpha} = \alpha \ket{\alpha}$$.

## 2. Measured Quantity

The Wigner function at the origin is related to the normally ordered exponential of the number operator:

$$
W_{\text{out}}(0) = \frac{2}{\pi} \langle{ : e^{-2 \hat{a}_{\text{out}}^\dagger \hat{a}_{\text{out}}} : }\rangle
$$

Substituting $\hat{a}_{\text{out}}$ , and applying normal ordering and coherent state properties, this becomes:

$$
W_{\text{out}}(0) = \frac{2}{\pi} \langle{ : \exp\left( -2T \left( \hat{a}_S - \sqrt{\frac{1 - T}{T}} \alpha \right)^\dagger \left( \hat{a}_S - \sqrt{\frac{1 - T}{T}} \alpha \right) \right) : }\rangle
$$

## 3. Identifying with the Quasiprobability Operator

This has the same structure as the \( s \)-parametrized quasidistribution function:

$$
W_S(\beta, s) = \frac{2}{\pi(1 - s)} \langle{ : \exp\left( -\frac{2}{1 - s} (\hat{a}^\dagger - \beta^*)(\hat{a} - \beta) \right) : }\rangle
$$

Comparing the forms, we identify:
$$
\beta = \sqrt{\frac{1 - T}{T}} \alpha
$$
$$
s = - \frac{1 - T}{T}
$$

Thus, the Wigner function (or more generally, the \( s \)-ordered quasidistribution) of the signal mode is:

$$
W_{\text{out}}(0) = \frac{1}{T} \, W_S\left( \sqrt{\frac{1 - T}{T}} \alpha ; - \frac{1 - T}{T} \right)
$$

## 4. Generalization to Arbitrary Probe State

For arbitrary probe state $\rho_P$ , the measured output is a convolution:

$$
W_{\text{out}}(0) = \int d^2 \beta \, W_S(\beta) \, W_P\left( \sqrt{\frac{T}{1 - T}} \beta \right)
$$

This shows that the experiment implements a filter-like measurement of the signal's phase space distribution.

## 5. Summary of Parameters

- $\alpha$: complex amplitude of the probe coherent state  both amplitude and phase determine the probed point in phase space.
- \( T \): beam splitter transmissivity  fixed per measurement; controls the ordering of the measured distribution.
- $\eta$: detector efficiency  affects the accuracy and resolution.

The experimental setup is based on a Mach-Zehnder-type interferometer designed to measure the Wigner function of a signal field using interference with a probe beam and photon counting. Below is the detailed explanation of each component and its role.

## 1. Laser Source
- **Type**: Frequency-stabilized, single-mode He:Ne laser.
- **Wavelength**: $\lambda = 632.8$ nm.
- **Polarization**: Linearly polarized in the plane of the setup.
- **Purpose**: Provides a stable, coherent beam that serves as the source for both signal and probe arms.

## 2. Beam Splitter BS1
- **Function**: Weakly reflective beam splitter that divides the laser beam:
  - Weakly reflected beam $\rightarrow$ Signal field.
  - Strong transmitted beam $\rightarrow$ Probe field.

## 3. Signal Field Preparation
- **Neutral Density (ND) Filter**: Reduces the intensity of the signal beam without altering its phase or polarization.
- **Mirror on Piezoelectric Translator (PZT)**: Introduces variable and controllable phase shifts, including simulated phase noise.
- **Purpose**: Creates pure and mixed coherent states of light for characterization.

## 4. Probe Field Control (Displacement Operation)
- **Half-Wave Plate (HWP)**: Rotates polarization to control intensity through a downstream polarizer. Rotation by angle $\theta$ leads to polarization rotated by $2\theta$.
- **Electro-Optic Modulator 1 (EOM1)**: A longitudinal Pockels cell that modifies the polarization-dependent intensity of the probe beam by altering the refractive index via an applied voltage.
- **Polarizer**: Oriented along the initial polarization axis; transmits light depending on polarization state after HWP and EOM1, achieving amplitude modulation.

### Amplitude Modulation Using HWP, EOM1, and Polarizer

The amplitude of the probe beam is controlled by adjusting its polarization and filtering it through a polarizer. This is achieved in three steps:

- **Half-Wave Plate (HWP)**:  
  Rotates the polarization of the probe beam by an angle $ 2\theta $, where $ \theta $ is the angle between the beams initial polarization and the fast axis of the HWP. This sets the polarization direction that enters EOM1.

- **Electro-Optic Modulator 1 (EOM1)**:  
  A Pockels cell that modifies the **phase difference** between orthogonal polarization components based on the applied voltage. This further adjusts the beam's polarization state in a voltage-controlled manner.

- **Polarizer**:  
  Positioned after EOM1, the polarizer transmits only the component of the light aligned with its axis. The combination of HWP + EOM1 determines how much of the probe beams intensity is transmitted through the polarizer, effectively modulating its **amplitude**.

Together, these elements allow precise, voltage-controlled modulation of the **probe beam's amplitude**, which in turn sets the **radial displacement** in phase space for Wigner function measurement.


## 5. Phase Control (Angular Coordinate in Phase Space)
- **Electro-Optic Modulator 2 (EOM2)**: Introduces phase modulation on the signal beam using the electro-optic effect. Equivalent to modulating probe phase, but more stable.
- **Purpose**: Allows scanning of angular coordinate $\omega$ in phase space: $\beta = \sqrt{n_{\text{vac}}} e^{i\omega}$.

## 6. Beam Splitter BS2
- **Type**: Nearly fully transmitting T = 98.6% .
- **Function**: Interferes the signal and probe beams. The signal beam undergoes displacement by weak mixing with the strong probe.

## 7. Aperture (A)
- **Function**: Blocks spurious reflections and unwanted modes from the interferometer output.

## 8. Photon Collection and Detection
- **Focusing Lens**: Focuses the displaced signal beam onto the active area of the photon detector.
- **Detector**: EG\&G SPCM-AQ-CD2749:
  - Silicon avalanche photodiode in Geiger mode.
  - Quantum efficiency $\eta > 70\% $.
  - Operated at low count rates to ensure single-photon resolution.

## 9. Data Acquisition and Control
- **Computer System**:
  - Records photon counts in time bins of $t = 30\,\mu s$, repeated over N = 8000 intervals.
  - Controls voltages applied to EOM1 and EOM2 for scanning amplitude and phase.

## 10. Measurement Principle
- The signal field is displaced by $-\beta$ , and the photon count distribution $p_n(\beta)$ is recorded.
- The Wigner function is reconstructed using:

$$
W(\beta) = \frac{2}{\pi} \sum_{n=0}^\infty (-1)^n p_n(\beta)
$$

- The variance in the measured value is given by:

$$
\text{Var}[P(\beta)] = \frac{1 - P(\beta)^2}{N}
$$

# 1. Measuring the Parity : Rahul Yadav

They define the parity $P(b)$ as:

$$
\Pi(\beta) = \sum_{n=0}^{\infty} (-1)^n p_n(\beta)
$$

where $p_n(b)$ is the measured probability of detecting $n$ photons at a given phase-space point $\beta$.

## Physical Meaning

This quantity $\Pi(\beta)$ is (apart from the overall constant $\frac{2}{\pi}$) equal to the Wigner function at that point:

$$
W(\beta) = \frac{2}{\pi} \Pi(\beta)
$$

## Experimental Procedure

At each point $\beta$, they collect $N$ counting intervals. Each interval is 30 ms long. Within each interval, exactly one of the outcomes $n$ photons detected occurs.

They build up the empirical distribution $p_n(\beta)$.

## Random Variable in Each Interval

Conceptually, in each 30 ms interval, you can think of measuring a random variable:

$$
X = (-1)^n
$$

This $X$ takes values $+1$ (if $n$ is even) or $-1$ (if $n$ is odd). Over many trials, its sample mean is precisely $P(b)$.



# 2. Error Formula

Once you view $\Pi(\beta)$ as the average of a $\pm 1$ variable over $N$ trials, the usual formula for the variance of the sample mean applies:

## Variance of a Single Trial

Since $X^2 = 1$ always (whether $X = +1$ or $X = -1$):

$$
\text{Var}[X] = \langle X^2 \rangle - \langle X \rangle^2 = 1 - \Pi(\beta)^2
$$

## Variance of the Sample Mean

If you have $N$ independent intervals (trials), the variance of the mean $\Pi(\beta)$ is:

$$
\text{Var}[\Pi(\beta)] = \frac{1 - \Pi(\beta)^2}{N}
$$

## Maximum Error

The largest statistical uncertainty occurs where $P(b) \approx 0$, because then $1 - \Pi(\beta)^2 \approx 1$. Hence, the maximum standard deviation is:

$$
\text{Var}[\Pi(\beta)] = \frac{1}{N}
$$

In the paper, $N = 8000$, so:

$$
\frac{1}{8000} \approx 1.1\%
$$

This is exactly what they quote as the peak statistical (sampling) error when $\Pi(\beta) = 0$.


# 3. Interpretation in the Paper

They collect $N = 8000$ intervals for each grid point in phase space. Compute $\Pi(\beta)$ by summing $(-1)^n p_n(b)$.

Propagate the error using:

$$
\text{Var}[\Pi(\beta)] = \frac{1 - \Pi(\beta)^2}{N}
$$

Hence, the quoted standard error near zero is $\frac{1}{N} \approx 1.1\%$.

Because the Wigner function $W(b)$ is simply $\frac{2}{\pi} \Pi(\beta)$, one would multiply the above standard deviation by $\frac{2}{\pi} \approx 0.6366$ if you want the absolute error in the Wigner function itself. But the paper often quotes the fractional error in $\Pi(\beta)$ directly, because that is what they measure first.

In short, the error formula is essentially the Bernoulli/Binomial sampling error for a random variable that takes values $\pm 1$ with probabilities even $n$ vs. odd $n$. That straightforwardly leads to:

$$
\text{Var}[\Pi(\beta)] = \frac{1 - \Pi(\beta)^2}{N}
$$

giving a maximum uncertainty of $\frac{1}{N}$ when $\Pi(\beta) = 0$.


## 4. Incorporating Experimental Imperfections

In practice, there are two major issues:  

### (a) Losses (Imperfect Photodetection)  
Losses arise due to the reflection from the beam splitter and the fact that the photodetector has a quantum efficiency $ \eta < 1 $.  

Prior theoretical work (Refs. [5,6] in the paper) shows that these losses modify the measured alternating sum so that it is not the ideal Wigner function but instead is proportional to a generalized quasidistribution $ W(\alpha; s) $.  

The ordering parameter $ s $ is given by  

$$
s =  - \frac{1-  \eta T}{\eta T}
$$  

which for our purposes tells us that losses essentially smear the measured distribution.  



### (b) Mode Mismatch  
Mode mismatch occurs because the probe (strong coherent beam used to perform the displacement) and the signal (the field under study) are never perfectly overlapping in all degrees of freedom.  

The overlap between the two normalized mode functions is quantified by $ \xi $ (with $ \xi = 1 $ for perfect overlap and $ \xi < 1 $ when there is mismatch).  

Importantly, the interference visibility $ v $ is related to $ \xi $ by  

$$
\xi \approx v /(2 - v)
$$  

so for high visibility (say, $ v \approx 98.5\% $), we have $ \xi \approx 97\% $.  

To include mode mismatch in the analysis, the probe mode is decomposed into two orthogonal parts:  

- The overlapping part, which interferes with the signal and contributes to the desired displacement.  
- The nonoverlapping part, which does not interfere and gives rise to independent Poissonian counts.  

The total photon count statistics is then a convolution of the statistics from these two components. In the language of probability, convolution in the photon number domain corresponds to a multiplication (or factorization) when one looks at the alternating sum.  



## 5. Factorization of the Alternating Sum  

A simple calculation (as the paper states) shows that the alternating sum from the convolution can be written as a product:  

$$
\Pi(\beta) = \underbrace{\exp \left( -\frac{1}{2}(1 - \xi) \, |\beta|^2 \right)}{\text{Mismatch factor}} \underbrace{\left[ \frac{\pi}{2\eta T} W(\beta; s) \right]}{\text{Ideal contribution}}
$$  

### Explanation:  
- *Mismatch Factor:* $ \exp \left( -\frac{1}{2}(1 - \xi) \, |\beta|^2 \right) $ arises because the interfering (overlapping) part of the probe field has its amplitude reduced by a factor of $ \xi $. The nonoverlapping part adds extra counts that are Poisson distributed. Combining these two contributions results in a multiplicative factor dependent on both the mismatch ($ \xi $) and the effective amplitude ($ \bar{b} $) of the probe field.  
- *Ideal Contribution:* $ \frac{\pi}{2 \eta T} W(\beta; s) $ would be the alternating sum youd obtain with perfect mode matching and detection but modified to include the effect of losses (through the ordering parameter $ s $).

- where

$$
W = \frac{2}{\pi} \exp \left( -2|\beta - \alpha_0|^2 \right)
$$



## 6. Specialization to a Coherent State  

For a coherent state $ |\alpha_0\rangle $, the ideal Wigner function is a Gaussian centered at $ \alpha_0 $. When you include the imperfections:  

- *Amplitude Attenuation:* The displacement provided by the probe is reduced by the factor $ \xi $ and further attenuated by the overall detection efficiency $ \eta T $. Thus, the center of the Gaussian becomes $ \xi \, \eta T \, \alpha_0 $ rather than $ \alpha_0 $.  
- *Multiplicative Constant:* The overall measured alternating sum is multiplied by a constant factor  

$$
\exp \left( -\frac{1}{2}(1 - \xi) \, \eta T |\alpha_0|^2 \right)
$$  

Therefore, Equation (4) takes the form:  

$$
\Pi(\beta) = \exp \left( -\frac{1}{2}(1 - \xi) \, \eta T |\alpha_0|^2 \right) \exp \left( -\frac{1}{2} \left| \beta - \sqrt{ \xi \, \eta \, T }
 \alpha_0 \right|^2 \right)
$$  

In words, this says that the measured alternating sum is a Gaussian function (as expected for a coherent state) whose center is shifted to the attenuated amplitude $ \xi \, \eta T \, \alpha_0 $ and whose height is reduced by the constant factor $ \exp \left( -\frac{1}{2}(1 - \xi) \, \eta T |\alpha_0|^2 \right) $.  

For the parameters used in the experiment, the paper gives:  

$$
\xi \approx 0.97, \quad \eta T |\alpha_0|^2 \approx 1.34
$$  

so the constant factor becomes:  

$$
\exp \left( -\frac{1}{2}(1 - 0.97) \times 1.34 \right) \approx \exp(-0.02) \approx 0.92
$$  

This explains why the measured Wigner function of the coherent state is slightly lower (about 92% of the ideal peak) than that of the vacuum state.

# Roshan - Simulations



```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import wigner, fock, coherent

# Parameters
N = 40
alpha0 = 1.0 
alpha0_ = -1.0 
grid_points = 100
xvec = np.linspace(-3, 3, grid_points)

# Vacuum and coherent states
vac = fock(N, 0)
coh = coherent(N, alpha0)

# Wigner function for vacuum and coherent
W_vac = wigner(vac, xvec, xvec)
W_coh = wigner(coh, xvec, xvec)

# Phase-diffused coherent state using sinusoidal modulation
mod_depth = 2.0
num_phases = 100
phases = mod_depth * np.sin(np.linspace(0, 2 * np.pi, num_phases))

W_diff = np.zeros((grid_points, grid_points))
for phi in phases:
    alpha_phi = alpha0_ * np.exp(1j * phi)
    psi_phi = coherent(N, alpha_phi)
    W_phi = wigner(psi_phi, xvec, xvec)
    W_diff += W_phi
W_diff /= num_phases

# 3D Plotting function for Wigner function
def plot_3d_surface(W, title, zlim=None):
    X, Y = np.meshgrid(xvec, xvec)
    fig = plt.figure(figsize=(6, 5))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, W, rstride=1, cstride=1,
                    facecolors=plt.cm.Greys(W / np.max(W)),
                    linewidth=0.3, edgecolor='black', antialiased=True, shade=False)
    ax.set_title(title)
    ax.set_xlabel("Re()")
    ax.set_ylabel("Im()")
    ax.set_zlabel("()")
    if zlim is not None:
        ax.set_zlim(zlim)
        
    ax.view_init(elev=30, azim=135)
    plt.tight_layout()
    plt.show()
    
# Plot function for 2D contour
def plot_bw_wigner(W, title, levels):
    X, Y = np.meshgrid(xvec, xvec)
    plt.figure(figsize=(4.5, 4))
    plt.contour(X, Y, W, levels=levels, colors='black')
    plt.xlabel("Re()")
    plt.ylabel("Im()")
    plt.title(title)
    plt.axis('square')
    plt.tight_layout()
    plt.show()
    

levels_std = np.arange(-0.3, 0.6, 0.05)    # Contour levels for all plots
levels_diff = np.linspace(0.05, 0.35, 30)  # More evenly spaced contour lines

# Plot all Wigner functions (2D Contours)
plot_bw_wigner(W_vac, "Vacuum State",levels=levels_std,)
plot_bw_wigner(W_coh, "Coherent State",levels=levels_std)
plot_bw_wigner(W_diff, "Phase-Diffused Coherent State",levels=levels_diff)

# Plot all Wigner functions (3D Surface)
plot_3d_surface(W_vac, "Vacuum State (3D Surface)")
plot_3d_surface(W_coh, "Coherent State (3D Surface)")
plot_3d_surface(W_diff, "Phase-Diffused Coherent State (3D Surface)", zlim=(0,0.25))
```

# Discussion of Wigner Function Plots

The Wigner function, a quasiprobability distribution in phase space, is computed for three different quantum states: vacuum, coherent, and phase-diffused coherent states. These states are visualized both in **2D contour plots** and **3D surface plots** to understand their phase space properties.

## 1. Vacuum State
The vacuum state, denoted as \$|0\rangle \$, is the ground state of the quantum harmonic oscillator. Its Wigner function exhibits the following features:
- The Wigner function for the vacuum state is **centered at the origin** of the phase space $\text{Re}(\beta) = 0, \text{Im}(\beta) = 0\$.
- The **distribution is symmetric** and has a **Gaussian shape**. It shows no oscillations or significant structure since the vacuum state is the least excited state with no coherent displacement.
- In the 2D contour plot, the Wigner function shows **circular contours**, indicating the **Gaussian nature** of the vacuum state.
- In the 3D surface plot, the Wigner function shows a smooth, **bell-shaped curve** centered at the origin, with the peak representing the highest probability density.

## 2. Coherent State
The coherent state, denoted as \$|\alpha\rangle \$, is a displaced state in phase space and is often associated with classical-like behavior in quantum mechanics. Its Wigner function has the following features:
- The Wigner function for the coherent state is **shifted from the origin**. The displacement is determined by the complex parameter \$\alpha_0 \$ in the real direction of phase space.
- The **distribution remains Gaussian-shaped**, but it is **shifted along the $ \text{Re}(\beta) \$ axis** due to the displacement of the state. This shift corresponds to the classical position of the harmonic oscillator in phase space.
- The 2D contour plot shows a **Gaussian distribution** that is displaced from the origin in the **real direction**.
- The 3D surface plot also shows a **shifted bell-shaped curve**, with the highest peak at the new center of the distribution. The plot provides a visual representation of the **displacement**.

## 3. Phase-Diffused Coherent State
The phase-diffused coherent state is generated by modulating the phase of a coherent state with a sinusoidal function. The key features of the Wigner function for this state are as follows:
- The **phase diffusion introduces oscillations** in the Wigner function, creating **two outer peaks**. These peaks correspond to the **turning points of the harmonically modulated phase**.
- The **contour plot** clearly shows these two outer peaks, which arise from the periodic phase shifts applied during the modulation. The **Wigner function is more spread out**, with the peaks positioned at certain distances from the origin in phase space.
- The **3D surface plot** reveals the distinct peaks more clearly, as the surface is **non-symmetric** and features **two prominent lobes** corresponding to the peaks.


## Summary
The Wigner function is a powerful tool for visualizing quantum states in phase space. The results obtained in this analysis show the following:
- The **vacuum state** has a simple Gaussian distribution centered at the origin.
- The **coherent state** introduces a displacement in phase space, leading to a shifted Gaussian distribution.
- The **phase-diffused coherent state** exhibits two outer peaks due to the phase modulation, which corresponds to the turning points of the harmonically modulated phase.

These results provide a comprehensive understanding of the behavior of different quantum states in phase space, as visualized by the Wigner function. The 2D contour and 3D surface plots offer complementary views of these distributions, making it easy to analyze the structure of the states.
